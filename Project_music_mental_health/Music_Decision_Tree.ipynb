```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'music.csv')
df.head()
```

### Data wrangling

```python
df = df.drop(columns = 'Timestamp')
```

```python
df = df.drop(columns = 'Permissions')
```

```python
df.isna().sum()
```

```python
df = df.dropna().copy()
```

```python
df = df.astype({'Age': 'int', 'BPM': 'int', 'Anxiety': 'int',  'Depression': 'int', 'Insomnia': 'int', 'OCD': 'int'})
```

```python
df.info()
```

```python
df.columns
```

```python
df['Music effects'].replace(['Improve','No effect', 'Worsen'],[1, 0, -1], inplace=True)
```

```python
cat_col = df.columns[10:26]
df[cat_col] = df[cat_col].apply(lambda x: x.astype('category'))
df[cat_col] = df[cat_col].apply(lambda x: x.cat.codes)
```

```python
df['Primary streaming service'] = df['Primary streaming service'].astype('category')
df['Primary streaming service'] = df['Primary streaming service'].cat.codes
df['Fav genre'] = df['Fav genre'].astype('category')
df['Fav genre'] = df['Fav genre'].cat.codes
```

```python
df.iloc[:, 3:6] = df.iloc[:, 3:6].replace({'Yes': 1, 'No': 0})
df.iloc[:, 7:9] = df.iloc[:, 7:9].replace({'Yes': 1, 'No': 0})
```

```python
df = df.query('BPM not in [0, 624, 999999999]')
```

```python
df.head()
```

```python
df.reset_index(drop = True)
```

### Machine Learning - Decision Tree

```python
x = df.drop(columns = 'Music effects')
x
```

```python
y = df['Music effects']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 27)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_mod = DecisionTreeClassifier(max_depth = 6, min_samples_split = 6, random_state = 27)
```

```python
mod = dt_mod.fit(x_train, y_train)
mod
```

```python
mod.score(x_train, y_train)
```

```python
mod.score(x_test, y_test)
```
