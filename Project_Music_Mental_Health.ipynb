#### About dataset
Music therapy, or MT, is the use of music to improve an individual's stress, mood, and overall mental health. MT is also recognized as an evidence-based practice, using music as a catalyst for "happy" hormones such as oxytocin.

However, MT employs a wide range of different genres, varying from one organization to the next.

The MxMH dataset aims to identify what, if any, correlations exist between an individual's music taste and their self-reported mental health. Ideally, these findings could contribute to a more informed application of MT or simply provide interesting sights about the mind.

**Definition of each mental health**  
    1. Anxiety: Anxiety disorders involve excessive fear or worry that differs from normal feelings of nervousness or anxiousness.      
    2. Depression: Depression is a mood disorder characterized by a persistent feeling of sadness and loss of interest.    
    3. Insomnia: Insomnia is a sleep disorder where individuals have trouble falling asleep, staying asleep, or waking up too early and not being able to go back to sleep.    
    4. OCD: Obsessive-Compulsive Disorder (OCD) is a disorder involving unwanted intrusive thoughts followed by feelings of anxiety, uncertainty, and even occasional panic.

**1. How many people are there in each age group, and what is the average number of hours they are listening?**

```python
def age_group(df2):
    if df2['Age'] >= 10 and df2['Age'] < 20:
        return '10-19'
    elif df2['Age'] >= 20 and df2['Age'] < 30:
        return '20-29'
    elif df2['Age'] >= 30 and df2['Age'] < 40:
        return '30-39'
    elif df2['Age'] >= 40 and df2['Age'] < 50:
        return '40-49'
    elif df2['Age'] >= 50 and df2['Age'] < 60:
        return '50-59'
    elif df2['Age'] >= 60:
        return '60+'
    
df2['Age_group'] = df2.apply(age_group, axis = 1)
```

```python
avg_hours = pd.DataFrame(df2.groupby('Age_group')['Hours per day'].agg(['count', 'mean'])).reset_index()\
            .rename(columns={'count': 'total_people', 'mean': 'avg_hour_listening'})

avg_hours
```

```python
sorted_values = avg_hours.sort_values('avg_hour_listening', ascending = False)['Age_group'].tolist()

plt.figure(figsize=(8,4))

sns.barplot(x='Age_group', y='avg_hour_listening', data = avg_hours, order = sorted_values, palette = 'Oranges');
plt.title('Average listening hours for each age group');
plt.ylabel('average hour');
```

**5. Mental Health and Music effects**

```python
# Distribution of mental health score
```

```python
# Music effect
```

```python
labels = ['Improve', 'No_effects', 'Worsen']
colors = ['#b78071', '#e7a87a', '#edbe7f']

Improve = 0
No_effects = 0
Worsen = 0

for i in df2['Music effects']:
    if i == 1:
        Improve += 1
    elif i == 0:
        No_effects += 1
    elif i == -1:
        Worsen += 1
        
plt.figure(figsize=(6, 6))  
plt.pie([Improve, No_effects, Worsen], labels = labels, colors = colors, autopct = '%.2f %%', textprops={'fontsize': 12} );
plt.title('Music effect ratio', fontdict = {'fontsize':15});
```

Approximately 75.61% of people with mental health conditions showed improvement from listening to music, 21.93% experienced no improvement, and only 2.45% worsened.

```python
effect = pd.DataFrame(df2.groupby('Fav genre').apply(
    lambda group: pd.Series({
        'total_people': group['Age'].count(),
        'Improve': (group['Music effects'] == 1).sum(),
        'No_effects': (group['Music effects'] == 0).sum(),
        'Worsen': (group['Music effects'] == -1).sum()})).sort_values('Improve', ascending = False).reset_index())

effect
```
