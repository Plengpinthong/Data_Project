 #### Description  
For this project, I performed machine learning using the Logistic Regression model from Scikit-learn to create a churn prediction model.

```python
import pandas as pd
```

```python
df_online_service = pd.read_csv(r'E:\Data_project\projrct\Data_Analyst_Project\Data_Analyst_Project\customer\archive (2)\Online_Services.csv')
df_payment = pd.read_csv(r'E:\Data_project\projrct\Data_Analyst_Project\Data_Analyst_Project\customer\archive (2)\Payment_Info.csv')
df_service = pd.read_csv(r'E:\Data_project\projrct\Data_Analyst_Project\Data_Analyst_Project\customer\archive (2)\Service_Options.csv')
df_status = pd.read_csv(r'E:\Data_project\projrct\Data_Analyst_Project\Data_Analyst_Project\customer\archive (2)\Status_Analysis.csv')
```

#### Data wrangling

```python
df_online_service['internet_type'] = df_online_service['internet_type'].astype('category')
df_online_service['internet_type'] = df_online_service['internet_type'].cat.codes
df_online_service.iloc[:, 1:11] = df_online_service.iloc[:, 1:11].replace({'Yes': 1, 'No': 0})
```

```python
df_online_service.head()
```

```python
df_payment = df_payment[['customer_id','contract', 'monthly_ charges', 'avg_monthly_long_distance_charges']]
df_payment.rename(columns={'monthly_ charges': 'monthly_charges', 'avg_monthly_long_distance_charges': 'monthly_long_distance_charges'}, inplace = True)
df_payment['contract'] = df_payment['contract'].astype('category')
df_payment['contract'] = df_payment['contract'].cat.codes
df_payment.head()
```

```python
df_service = df_service[['customer_id', 'tenure', 'unlimited_data', 'offer' ]]
df_service.rename(columns={'tenure':'tenure(month)'}, inplace = True)
df_service.iloc[:, 2] = df_service.iloc[:, 2].replace({'Yes': 1, 'No': 0})
df_service['offer'] = df_service['offer'].astype('category')
df_service['offer'] = df_service['offer'].cat.codes
df_service.head()
```

```python
df_status = df_status[['customer_id', 'satisfaction_score', 'churn_score', 'churn_label']]
df_status['churn_label'] = df_status['churn_label'].astype('category')
df_status['churn_label'] = df_status['churn_label'].cat.codes
df_status.head()    
```

```python
# merge all tables

df = pd.merge(pd.merge(pd.merge(df_online_service, df_payment, on='customer_id'), df_service, on='customer_id'), df_status, on='customer_id')
df = df.drop(columns = 'customer_id')
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.drop(columns = 'churn_label')
x
```

```python
y = df['churn_label']
y
```

#### Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 61)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
x_train_scale = scale.fit_transform(x_train)
x_test_scale = scale.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(random_state = 0).fit(x_train_scale, y_train)
```

```python
y_pred = log_model.predict(x_train_scale)
y_result = pd.DataFrame(y_pred)
y_result
```

```python
# score of train data
log_model.score(x_train_scale, y_train)
```

```python
# score of test data
log_model.score(x_test_scale, y_test)
```
